## Feature Importance
Random Forests assign equal or similar importance to features that are highly correlated. In addition, when features are correlated, the importance assigned is lower than the importance attributed to the feature itself, should the tree be built without the correlated counterparts.

Therefore, instead of eliminating features based on importance by brute force like we did in the previous notebook, we could get a better selection by removing one feature at a time, and recalculating the importance on each round. This procedure is called Recursive Feature Elimination (RFE)

RFE is a hybrid between embedded and wrapper methods: it is based on computation derived when fitting the model, but it also requires fitting several models.s.**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import roc_auc_score

sel_ = RFE(RandomForestClassifier(n_estimators=10, random_state=42), n_features_to_select=12)
sel_.fit(X_train, y_train)
```

```python
# selected features
selected_feat = X_train.columns[(sel_.get_support())]
len(selected_feat)
```

```python
# let's display the list of features
selected_feat
```

```python
# Function to build Random Forests and compare performance in train and test sets
def run_randomForests(X_train, X_test, y_train, y_test):
    # Initialize the Random Forest classifier with specified parameters
    rf = RandomForestClassifier(n_estimators=200, random_state=39, max_depth=4)
    
    # Fit the model on training data
    rf.fit(X_train, y_train)
    
    # Evaluate performance on the train set
    print('Train set:')
    pred_train = rf.predict_proba(X_train)  # Get predicted probabilities
    print('Random Forests roc-auc: {:.4f}'.format(roc_auc_score(y_train, pred_train, multi_class='ovr')))
    
    # Evaluate performance on the test set
    print('Test set:')
    pred_test = rf.predict_proba(X_test)  # Get predicted probabilities
    print('Random Forests roc-auc: {:.4f}'.format(roc_auc_score(y_test, pred_test, multi_class='ovr')))
```

### With 12 variables

```python
# features selected recursively
run_randomForests(X_train[selected_feat],
                  X_test[selected_feat],
                  y_train, y_test)
```

### With all variables

```python
# features selected recursively
run_randomForests(X_train,
                  X_test,
                  y_train, y_test)
```

## Model selection

### Random Forest

```python
from sklearn.model_selection import StratifiedKFold, cross_validate

# Initialize the Random Forest classifier with specified parameters
rf = RandomForestClassifier(n_estimators=200, random_state=39, max_depth=4)

# Leave P Out Cross-Validation using StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=4)

# Estimate generalization error using cross-validation
clf = cross_validate(
    rf,
    X_train,
    y_train,
    scoring='accuracy',
    return_train_score=True,
    cv=skf,  # k-fold cross-validation
)
# Fit the model on the entire training data (to evaluate final performance on the test set)
rf.fit(X_train, y_train)
```

```python
# Get the hyperparameters
print(rf.get_params())
```

## Model Evaluation

```python
# Display the number of test scores (should be equal to the number of folds)
print('Number of Test Scores:', len(clf['test_score']))
print('mean train set accuracy: ', np.mean(
    clf['train_score']), ' +- ', np.std(clf['train_score']))
print('mean test set accuracy: ', np.mean(
    clf['test_score']), ' +- ', np.std(clf['test_score']))
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix, classification_report

# Evaluate performance on the train set
print('Train set:')
pred_train = rf.predict(X_train)  # Get predicted classes
pred_train_proba = rf.predict_proba(X_train)  # Get predicted probabilities
print('Train Accuracy: {:.4f}'.format(accuracy_score(y_train, pred_train)))
print('Train ROC AUC: {:.4f}'.format(roc_auc_score(y_train, pred_train_proba, multi_class='ovr')))

# Evaluate performance on the test set
print('Test set:')
pred_test = rf.predict(X_test)  # Get predicted classes
pred_test_proba = rf.predict_proba(X_test)  # Get predicted probabilities
print('Test Accuracy: {:.4f}'.format(accuracy_score(y_test, pred_test)))
print('Test ROC AUC: {:.4f}'.format(roc_auc_score(y_test, pred_test_proba, multi_class='ovr')))

# Confusion matrix for the test set
cm = confusion_matrix(y_test, pred_test)
print('Confusion Matrix:')
print(cm)

# Classification report
print('Classification Report:')
print(classification_report(y_test, pred_test))
```

## Hyperparameter Optimization

```python
import optuna
from sklearn.model_selection import cross_val_score
# the objective function takes the hyperparameter space
# as input

def objective(trial):

    rf_n_estimators = trial.suggest_int("rf_n_estimators", 100, 1000)
    rf_criterion = trial.suggest_categorical("rf_criterion", ['gini', 'entropy'])
    rf_max_depth = trial.suggest_int("rf_max_depth", 1, 4)
    rf_min_samples_split = trial.suggest_float("rf_min_samples_split", 0.01, 1)
    
    model = RandomForestClassifier(
        n_estimators=rf_n_estimators,
        criterion=rf_criterion,
        max_depth=rf_max_depth,
        min_samples_split=rf_min_samples_split,
    )

    score = cross_val_score(model, new_x_train, y_train, cv=3)
    accuracy = score.mean()
    return accuracy
```

```python
study = optuna.create_study(
    direction="maximize",
    sampler=optuna.samplers.TPESampler(
        # the number of initial points to sample at random before doing the sequential search
        n_startup_trials=10
    ),
)
study.optimize(objective, n_trials=20)
```
